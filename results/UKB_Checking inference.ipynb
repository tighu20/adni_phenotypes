```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
disease_info_df = pd.read_csv('../data/ukb_diagnosis.csv', index_col=0)
disease_info_df.shape
```

```python
np.unique(disease_info_df['diag_icd10'], return_counts=True)
```

```python
all_results_df = pd.read_csv('latest_output_ukb_50.csv', index_col=0)
all_results_df
```

```python
joined_df = all_results_df.join(disease_info_df, how='inner')
joined_df
```

```python
np.unique(joined_df.diag_icd10, return_counts=True)
```

```python
all_data = [joined_df[joined_df.diag_icd10 == 'AD']['mean'].values, 
           joined_df[joined_df.diag_icd10 == 'AD']['std'].values,
           joined_df[joined_df.diag_icd10 == 'PD']['mean'].values,
           joined_df[joined_df.diag_icd10 == 'PD']['std'].values]


_ = plt.boxplot(all_data, labels=['AD mean', 'AD std', 'PD mean', 'PD std'])
```
